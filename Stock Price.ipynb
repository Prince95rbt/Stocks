```python
import pandas as pd
import numpy as np
```

```python
df1 = pd.read_csv("D:\\Python Preparatory\\Assignment\\Stock_File_1.csv", parse_dates = True)
df1
```

```python
df2 = pd.read_csv("D:\\Python Preparatory\\Assignment\\Stock_File_2.txt", sep = ',',parse_dates = True)
df2
```

```python
stock = pd.concat([df1,df2], ignore_index = True)
```

```python
stock
```

```python
stock.isnull().sum()
```

```python
stock.dropna(inplace = True, axis = 0)
```

```python
stock.parse_dates = True
```

```python
stock
```

```python
stock['Date'] = pd.to_datetime(stock['Date'])
```

```python
stock['Date'].max() - stock['Date'].min()
```

```python
stock.set_index('Date',inplace=True)
```

```python
stock.iloc[0:].describe().astype(int)
```

```python
stock['Volume'] = pd.to_numeric(stock['Volume'], errors = 'coerce')
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
stock
```

```python
#market capitalisation
stock['MarketCap'] = stock['Open'] * stock['Volume']
```

```python
stock.describe()
```

```python
#Percentage increase in stock value
stock['returns'] = (stock['Close']/stock['Close'].shift(1)) -1
stock['returns'].hist(bins = 100, label = 'stock_price', alpha = 0.5, figsize = (15,7))
```

```python
stock['Volume'].plot(figsize = (15,7))
```

```python
stock['Close'].plot(label = 'Stocks at Closing', figsize = (15,7))
stock['Open'].plot(label = 'Stock Price at Open')
stock['High'].plot(label = 'Stock Price at High')
stock['Low'].plot(label = 'Stock Price at Low')
plt.legend()
plt.title('Stock Prices')
plt.show()
```
